```python
# Task1 Exploring and visualizing dataset
# By: Saif Ullah
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================
# STEP 1: LOAD THE DATASET
# ============================================
# Load Iris dataset using seaborn
iris = sns.load_dataset('iris')
print("Dataset loaded successfully!")

# ============================================
# STEP 2: BASIC DATA INSPECTION
# ============================================
# Print shape of dataset
print(f"\nDataset Shape: {iris.shape}")

# Print column names
print(f"\nColumn Names: {list(iris.columns)}")

# Print first few rows
print(f"\nFirst 5 rows:")
print(iris.head())

# ============================================
# STEP 3: SUMMARY STATISTICS
# ============================================
# Use info() for dataset information
print(f"\nDataset Info:")
print(iris.info())

# Use describe() for statistical summary
print(f"\nStatistical Summary:")
print(iris.describe())

# ============================================
# STEP 4: DATA VISUALIZATION
# ============================================
# Create figure with subplots
plt.figure(figsize=(15, 10))

# 4.1 Scatter plot - Show relationships between features
plt.subplot(2, 3, 1)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')

plt.subplot(2, 3, 2)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width')

# 4.2 Histograms - Show value distributions
plt.subplot(2, 3, 3)
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    plt.hist(subset['sepal_length'], alpha=0.5, label=species)
plt.title('Sepal Length Distribution')
plt.legend()

plt.subplot(2, 3, 4)
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    plt.hist(subset['petal_length'], alpha=0.5, label=species)
plt.title('Petal Length Distribution')
plt.legend()

# 4.3 Box plots - Identify outliers
plt.subplot(2, 3, 5)
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Sepal Length Box Plot')

plt.subplot(2, 3, 6)
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title('Petal Length Box Plot')

plt.tight_layout()
plt.show()
```
